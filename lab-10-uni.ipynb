```python
class TreeNode:
    def __init__(self, name, fuel, selling_price):
        self.name = name
        self.fuel = fuel
        self.price = selling_price
        self.children = []

    def add_child(self, child_node):
        self.children.append(child_node)

all_petrol_list_cars = []

def build_car_tree(diction, petrol_root,diesel_root):
    for item in diction:
        if item["fuel"] == "Petrol":
            # Create a new root node if the tree is empty
            if petrol_root is None:
                
                petrol_root = TreeNode(item["name"], item["fuel"], item["selling_price"])
#                 print("petrol_node")
            else:
                # Add a child node to the existing tree
                petrol_root.add_child(TreeNode(item["name"], item["fuel"], item["selling_price"]))
#                 print("petrol_child_node")

        elif item["fuel"] == "Diesel":
            if diesel_root is None:
                diesel_root = TreeNode(item["name"], item["fuel"], item["selling_price"])
#                 print("diese_root")
            else:
                # Add a child node to the existing tree
                diesel_root.add_child(TreeNode(item["name"], item["fuel"], item["selling_price"]))
#                 print("dieses_child")
        else:
            print("sdfa")
    
    return petrol_root,diesel_root


features=df.head(5).to_dict(orient='records')
features
petrol_node = None
diesel_node=None

petrol_root,diesel_root = build_car_tree(features, petrol_node,diesel_node)
    
def dfs(node):
    print(node.name,end="is of ")
    print(node.price)
    for child in node.children:
        dfs(child)
    

print("Diesel nodes are \n:")
dfs(diesel_root)
print(f"petrol nodes are \n ")
dfs(petrol_root)

```
